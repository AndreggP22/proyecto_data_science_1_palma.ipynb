```python
# Reimportar librerías tras reinicio del entorno
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.feature_selection import SelectKBest, f_regression
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Limpieza y conversión
df['Price'] = df['Price'].astype(str).str.replace(',', '.')
df['Price'] = pd.to_numeric(df['Price'], errors='coerce')
df['ABV'] = df['ABV'].astype(str).str.replace(',', '.')
df['ABV'] = pd.to_numeric(df['ABV'], errors='coerce')
df['Capacity_num'] = df['Capacity'].str.replace('ML', '', regex=False)
df['Capacity_num'] = pd.to_numeric(df['Capacity_num'], errors='coerce')

# Filtrar nulos
df_clean = df.dropna(subset=['Price', 'ABV', 'Type', 'Closure'])

# Codificación de variables categóricas
data = df_clean.copy()
label_cols = ['Type', 'Closure', 'Country', 'Style']
label_encoders = {}

for col in label_cols:
    le = LabelEncoder()
    data[col] = le.fit_transform(data[col])
    label_encoders[col] = le

# Definir X y y
X = data.drop(columns=[
    'Price', 'Title', 'Characteristics', 'Per', 'Per bottle / case / each',
    'Grape', 'Secondary Grape Varieties', 'Region', 'Appellation',
    'Capacity', 'Unit'
], errors='ignore')
y = data['Price']

# Escalado
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Feature selection
selector = SelectKBest(score_func=f_regression, k=5)
X_selected = selector.fit_transform(X_scaled, y)
selected_features = X.columns[selector.get_support()].tolist()

# Modelo y evaluación
X_train, X_test, y_train, y_test = train_test_split(X_selected, y, test_size=0.2, random_state=42)
model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

# Métricas
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

(selected_features, mae, mse, r2)



```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x=y_test, y=y_pred, alpha=0.7, color='mediumseagreen')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', lw=2)
plt.title('Regresión Lineal: Precio Real vs Precio Predicho')
plt.xlabel('Precio Real (€)')
plt.ylabel('Precio Predicho (€)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

La línea roja punteada representa una predicción perfecta (donde el valor real = predicho).

Los puntos dispersos reflejan que muchas predicciones no están perfectamente alineadas, lo cual es coherente con el R² bajo (~0.165) obtenido.

Aun así, existe una tendencia ascendente leve, indicando que el modelo captura algunas relaciones útiles.

Se utilizó el método **SelectKBest** con la función **f_regression**:

Objetivo: seleccionar las variables numéricas más correlacionadas con la variable objetivo (Price)

Se aplicó un modelo de **Regresión Lineal**:

Entrenamiento: sobre el 80% de los datos

Prueba/validación: con el 20% restante

Variables predictoras: las seleccionadas por SelectKBest

Este modelo fue elegido por su interpretabilidad y adecuación para relaciones lineales entre variables continuas.
